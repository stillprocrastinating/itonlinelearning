```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('0_telecom.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Check summary statistics
print(df.describe())
```

```python
# Check data types and missing values
print(df.info())
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python
# Visualise the distribution of customer churn
sns.countplot(x='churn', data=df)
plt.show()
```

```python
# Visualise the relationship between customer tenure and churn
sns.boxplot(x='churn', y='account length', data=df)
plt.show()
```

```python
# Visualise the distribution of daily charges for churned and non-churned customers
sns.histplot(x='total day charge', hue='churn', data=df, kde=True)
plt.show()
```

```python
# Visualise the relationship between tenure and daily charges
sns.scatterplot(x='account length', y='total day charge', hue='churn', data=df)
plt.show()
```
