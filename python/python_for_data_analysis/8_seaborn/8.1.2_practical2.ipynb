```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('0_housing.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Check summary statistics
print(df.describe())
```

```python
# Check data types and missing values
print(df.info())
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# Set seaborn style
sns.set(style="ticks")
```

```python
# Visualise the distribution of housing prices
sns.histplot(x='Price', data=df, kde=True)
plt.show()
```

```python
# Visualise the relationship between living area size and sale price
sns.scatterplot(x='SqFt', y='Price', data=df)
plt.show()
```

```python
# Visualise the distribution of sale prices by the number of bedrooms
sns.boxplot(x='Bedrooms', y='Price', data=df)
plt.show()
```

```python
# Visualise the relationship between the number of offers and sale price
sns.barplot(x='Offers', y='Price', data=df)
plt.show()
```
