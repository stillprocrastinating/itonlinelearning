```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('0_restaurant.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Check summary statistics
print(df.describe())
```

```python
# Check data types and missing values
print(df.info())
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python
# Visualise the distribution of review ratings
sns.countplot(x='Rating', data=df)
plt.show()
```

```python
# Visualise the distribution of review lengths
df['Review_str'] = str(df['Review'])
df['Review_len'] = df['Review_str'].apply(len)
sns.histplot(x='Review_len', data=df, kde=True)
plt.show()
```

```python
# Visualise the relationship between review length and rating
sns.boxplot(x='Rating', y='Review_len', data=df)
plt.show()
```

```python
# Visualise the distribution of ratings over time
df['Time'] = pd.to_datetime(df['Time'])
sns.lineplot(x='Time', y='Rating', data=df)
plt.xticks(rotation=45)
plt.show()
```
